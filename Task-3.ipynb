```python
#Importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Reading dataset
df=pd.read_csv("SampleSuperstore.csv")
```

```python
df.head()
```

```python
#Checking for type error
df.info()
```

```python
df.describe()
```

```python
#Checking for null values
df.isna().sum()
```

```python
df.columns=['Ship_mode','Segment','Country','City','State','Postal_code','Region','Category','Sub-category','Sales','Quantity','Discount','Profit']
```

```python
df.head()
```

```python
#Subsetting dataframe by Profit>0
df_new=df[df['Profit']>0]
```

```python
#Sold Quantity & Gained Profit based on Category and Sub-Category
df_new.groupby(['Category','Sub-category'])["Quantity",'Profit'].agg(sum).plot.bar(figsize=(8,4))
plt.ylabel("Quantity")
plt.title("Sold quantities by Category & Sub-Category")
plt.show()
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
#State-wise Profit and Sales
df_new.groupby(['State'])['Profit','Sales'].sum().plot.bar(figsize=(10,6))
plt.show()
```

```python
#Region-wise Profit and Sales
df.groupby(['Region'])['Profit','Sales'].sum().plot.bar(figsize=(10,6))
plt.show()
```

```python
#Category-wise division of Regions
plt.figure(figsize=(20,15))
plt.subplot(2,2,1)
sns.countplot('Region',data=df,hue='Category')
plt.title('Category-wise division of Regions')
#Sub-category-wise devision of Category
plt.figure(figsize=(20,15))
plt.subplot(2,2,2)
sns.countplot('Category',data=df,hue='Sub-category')
plt.legend(loc=[1,0])
plt.title('Sub-category-wise devision of Category')
#Countplot of Sub-category
plt.figure(figsize=(20,15))
plt.subplot(2,2,3)
sns.countplot('Sub-category',data=df)
plt.xticks(rotation=90)
#countplot of States
plt.figure(figsize=(20,15))
plt.subplot(2,2,4)
sns.countplot('State',data=df)
plt.xticks(rotation=90)
plt.show()
```
